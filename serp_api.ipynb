# Clusters

```python
scholar_profile = json.load(open("scholar_profile.json", "r"))
```

```python
def extract_cluster(url):
    return urllib.parse.parse_qs(urllib.parse.urlparse(url).query)["cluster"][0]


for article in scholar_profile["articles"]:
    print(article["title"])
    article["link_list"] = []
    for scholar_article in article["citation"]["scholar_articles"]:
        cluster_id = extract_cluster(scholar_article["link"])
        print("\t", cluster_id)

        result = client.search(
            {
                "engine": "google_scholar",
                "hl": "en",
                "cluster": cluster_id,
                "num": "100",
            }
        )
        for org in result.as_dict()["organic_results"]:
            if "link" in org.keys():
                article["link_list"].append(org["link"])
                print("\t\t", org["link"])

json.dump(scholar_profile, open("scholar_profile.json", "w"))
```
